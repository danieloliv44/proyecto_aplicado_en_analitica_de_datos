```python
import os
from pathlib import Path
import pandas as pd
import numpy as np
```

```python
# Obtener la raíz del proyecto (sube dos niveles desde notebooks/)
PROJECT_ROOT = Path().resolve().parents[1]
# Construir la ruta al archivo
data_path = PROJECT_ROOT / "data" / "raw" / "Municipios.xlsx"
print(f"Ruta del archivo: {data_path}")
```

```python
Municipios = pd.read_excel(data_path, sheet_name="Municipios")
Municipios.head()
```

```python
Municipios.drop(columns=["OBJECTID", "MpArea", "MpNorma", "MpCategor", "Restriccion", "SHAPE_Length", "SHAPE_Area"], inplace=True)
Municipios.head()
```

```python
# Librerías 
import unicodedata

Caldas = Municipios[Municipios['Depto'] == 'Caldas'].copy()

# Renombrar columna
Caldas = Caldas.rename(columns={'Area_CTM': 'AREA_HA'})

# Función para quitar tildes
def quitar_tildes(texto):
    return ''.join(
        c for c in unicodedata.normalize('NFD', texto)
        if unicodedata.category(c) != 'Mn'
    )

Caldas['DEPARTAMENTO'] = Caldas['Depto'].str.upper()
Caldas['MUNICIPIO'] = Caldas['MpNombre'].apply(lambda x: quitar_tildes(x).upper())
```

```python
PROJECT_ROOT = Path().resolve().parents[1]
output_path = PROJECT_ROOT / "data" / "processed" / "municipios.csv"
print(output_path)
Caldas.to_csv(output_path, index=False, encoding='utf-8-sig')
```
