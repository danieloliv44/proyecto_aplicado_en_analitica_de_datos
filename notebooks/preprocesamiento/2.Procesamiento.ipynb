```python
from pathlib import Path
import pandas as pd

# Raíz del proyecto: sube dos niveles desde notebooks/preprocesamiento
PROJECT_ROOT = Path.cwd().parents[1]

ruta_clima = PROJECT_ROOT / "data" / "raw" / "clima_mensual_caldas_2005_2025.csv"

clima = pd.read_csv(ruta_clima)
clima.head()
```

```python
# Tipificación y variables de fecha
clima["date"] = pd.to_datetime(clima["date"])

clima["anio"] = clima["date"].dt.year
clima["mes"] = clima["date"].dt.month

clima["municipio"] = clima["municipio"].astype("category")

print("Dimensiones:", clima.shape)
print("Municipios:", clima["municipio"].nunique())
print("Rango fechas:", clima["date"].min(), "→", clima["date"].max())

clima.head()
```

Se cargó la base climática mensual extraída desde Google Earth Engine y se ajustaron los tipos de datos principales. La fecha se convirtió a formato datetime, se derivaron las columnas de año y mes, y el municipio se tipificó como categoría para facilitar el procesamiento posterior.

```python
# Renombrar variables climáticas para facilitar interpretación
clima = clima.rename(columns={
    "temperature_2m": "temp_media_k",
    "temperature_2m_min": "temp_min_k",
    "temperature_2m_max": "temp_max_k",
    "dewpoint_temperature_2m": "temp_rocio_k",
    "total_precipitation_sum": "precip_m",
    "volumetric_soil_water_layer_1": "humedad_suelo_capa1",
    "volumetric_soil_water_layer_2": "humedad_suelo_capa2",
    "surface_solar_radiation_downwards_sum": "radiacion_solar_j_m2",
    "potential_evaporation_sum": "evap_potencial_m",
    "total_evaporation_sum": "evap_real_m",
    "surface_runoff_sum": "escorrentia_superficial_m",
    "sub_surface_runoff_sum": "escorrentia_subsuperficial_m",
    "u_component_of_wind_10m": "viento_u_ms",
    "v_component_of_wind_10m": "viento_v_ms",
    "surface_pressure": "presion_superficie_pa"
})

clima.head()
```

Se renombraron las variables climáticas con nombres más interpretables, manteniendo referencia a sus unidades originales. Esto facilita la lectura del dataset y la posterior construcción de variables derivadas para el modelo.

```python
# Conversión de unidades

# Temperatura: Kelvin a Celsius
clima["temp_media_c"] = clima["temp_media_k"] - 273.15
clima["temp_min_c"] = clima["temp_min_k"] - 273.15
clima["temp_max_c"] = clima["temp_max_k"] - 273.15
clima["temp_rocio_c"] = clima["temp_rocio_k"] - 273.15

# Precipitación: metros a mm
clima["precip_mm"] = clima["precip_m"] * 1000

# Evapotranspiración: metros a mm
clima["et_potencial_mm"] = clima["evap_potencial_m"] * 1000
clima["et_real_mm"] = clima["evap_real_m"] * 1000

# Escorrentía: metros a mm
clima["escorrentia_sup_mm"] = clima["escorrentia_superficial_m"] * 1000
clima["escorrentia_sub_mm"] = clima["escorrentia_subsuperficial_m"] * 1000

clima.head()
```

Se realizó la conversión de unidades de las variables climáticas. Las temperaturas fueron transformadas de Kelvin a grados Celsius, mientras que las variables de precipitación, evapotranspiración y escorrentía se convirtieron de metros a milímetros para asegurar  coherencia de las variables dentro del modelo.

#### vairables derivadas 

```python
clima_features = clima.copy()

# Balance hídrico: precipitación menos evapotranspiración potencial
clima_features["balance_hidrico_mm"] = (
    clima_features["precip_mm"] - clima_features["et_potencial_mm"]
)

# Amplitud térmica
clima_features["amplitud_termica_c"] = (
    clima_features["temp_max_c"] - clima_features["temp_min_c"]
)

# Velocidad del viento a partir de componentes U y V
clima_features["velocidad_viento_ms"] = (
    (clima_features["viento_u_ms"]**2 + clima_features["viento_v_ms"]**2) ** 0.5
)

clima_features[[
    "municipio", "date", "precip_mm", "et_potencial_mm",
    "balance_hidrico_mm", "temp_min_c", "temp_max_c",
    "amplitud_termica_c", "velocidad_viento_ms"
]].head()
```

Se construyeron variables derivadas para representar mejor las condiciones agroclimáticas. El balance hídrico aproxima condiciones de déficit o exceso de agua, la amplitud térmica resume la diferencia entre temperatura máxima y mínima, y la velocidad del viento integra los componentes horizontal y vertical del viento en una sola medida 

```python
# Variables derivadas adicionales

# Déficit hídrico (similar a balance pero enfocado en sequía)
clima_features["deficit_hidrico_mm"] = (
    clima_features["et_potencial_mm"] - clima_features["precip_mm"]
)

# Indicador de sequía (binario simple)
clima_features["estres_hidrico"] = (
    clima_features["balance_hidrico_mm"] < 0
).astype(int)

# Relación ET real vs potencial
clima_features["ratio_et"] = (
    clima_features["et_real_mm"] / (clima_features["et_potencial_mm"] + 1e-6)
)

# Humedad total del suelo
clima_features["humedad_suelo_total"] = (
    clima_features["humedad_suelo_capa1"] +
    clima_features["humedad_suelo_capa2"]
)

# Diferencia térmica con punto de rocío (proxy humedad atmosférica)
clima_features["dif_temp_rocio"] = (
    clima_features["temp_media_c"] - clima_features["temp_rocio_c"]
)

# Radiación (mantener pero con nombre más corto)
clima_features["radiacion"] = clima_features["radiacion_solar_j_m2"]

# Presión en kPa (más interpretable)
clima_features["presion_kpa"] = clima_features["presion_superficie_pa"] / 1000

clima_features.head()
```

Se construyeron variables derivadas adicionales orientadas a capturar condiciones agroclimáticas relevantes. Estas incluyen indicadores de déficit hídrico, estrés por falta de agua, relación entre evapotranspiración real y potencial, humedad total del suelo y variables asociadas a la dinámica térmica y atmosférica. 

```python
clima_features = clima_features.sort_values(
    ["municipio", "date"]
).copy()
```

```python
# Acumulados de lluvia
clima_features["precip_3m"] = (
    clima_features.groupby("municipio")["precip_mm"]
    .rolling(3).sum()
    .reset_index(level=0, drop=True)
)

clima_features["precip_6m"] = (
    clima_features.groupby("municipio")["precip_mm"]
    .rolling(6).sum()
    .reset_index(level=0, drop=True)
)


# Balance hídrico acumulado

clima_features["balance_3m"] = (
    clima_features.groupby("municipio")["balance_hidrico_mm"]
    .rolling(3).sum()
    .reset_index(level=0, drop=True)
)

clima_features["balance_6m"] = (
    clima_features.groupby("municipio")["balance_hidrico_mm"]
    .rolling(6).sum()
    .reset_index(level=0, drop=True)
)
```

Se incorporaron variables acumuladas y rezagos temporales con el fin de capturar efectos dinámicos del clima sobre el cultivo. Los acumulados permiten representar condiciones climáticas recientes, mientras que los rezagos incorporan memoria temporal,esto ya quelos impactos del clima no son inmediatos.

```python
print("Dimensiones:", clima_features.shape)
print("Municipios:", clima_features["municipio"].nunique())
print("Rango:", clima_features["date"].min(), "→", clima_features["date"].max())
print("Nulos:\n", clima_features.isna().sum())
```

#### Lectura NDVI

```python
ruta_ndvi = PROJECT_ROOT / "data" / "raw" / "ndvi_mensual_caldas_2005_2025.csv"

ndvi = pd.read_csv(ruta_ndvi)

# Limpiar columnas innecesarias
ndvi = ndvi.drop(columns=["system:index", ".geo"], errors="ignore")

# Tipificar
ndvi["date"] = pd.to_datetime(ndvi["date"])
ndvi["anio"] = ndvi["date"].dt.year
ndvi["mes"] = ndvi["date"].dt.month
ndvi["municipio"] = ndvi["municipio"].astype("category")

ndvi.head()
```

```python
print("Dimensiones:", ndvi.shape)
print("Municipios:", ndvi["municipio"].nunique())
print("Rango:", ndvi["date"].min(), "→", ndvi["date"].max())
print("Nulos:\n", ndvi.isna().sum())
```

#### MERGE

```python
df_model = clima_features.merge(
    ndvi[["municipio", "date", "ndvi_mean"]],
    on=["municipio", "date"],
    how="inner"
)

df_model.head()
```

```python
print("Dimensiones:", df_model.shape)
print("Municipios:", df_model["municipio"].nunique())
print("Rango fechas:", df_model["date"].min(), "→", df_model["date"].max())

print("Nulos:\n", df_model.isna().sum().sort_values(ascending=False))
```
