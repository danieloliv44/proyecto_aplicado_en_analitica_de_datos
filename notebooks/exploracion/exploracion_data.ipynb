```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path
```

```python
# Raíz del proyecto
PROJECT_ROOT = Path().resolve()
# Ruta a la carpeta de datos procesados
path = PROJECT_ROOT / "data" / "processed"
df = pd.read_csv(path / 'Dataset_Consolidado.csv')
```

```python
print(f"Municipios: {df['Municipio'].nunique()}")
print(f"Años: {df['Año'].min()}-{df['Año'].max()}")
print(f"Valores nulos: {df.isnull().sum().sum()}")
```

```python
df.describe()
```

```python
df.groupby('Municipio')['Rendimiento (t/ha)'].mean().sort_values(ascending=False)
```

```python
import matplotlib.pyplot as plt

df.groupby('Año')['Rendimiento (t/ha)'].mean().plot(figsize=(20, 5))
plt.title('Rendimiento Promedio por Año')
plt.xlabel('Año')
plt.ylabel('Rendimiento (t/ha)')
plt.grid(alpha=0.3)
plt.show()
```
